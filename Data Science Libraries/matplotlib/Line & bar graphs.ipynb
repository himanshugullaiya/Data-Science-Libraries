#### Plotting Line & Bar Graphs

```python
x = np.arange(0,5)
y_sq = x**2
y_cu = x**3
y_li = x
```

```python
# format strings
# fmt = '[marker][line][color]'  'o-g', '-g' https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html
```

```python
plt.plot(x,y_li, color = 'orange', linestyle = '--', linewidth = 3,  label = 'Linear')
plt.plot(x,y_sq, color = '#adad3b',linestyle = ':', marker = 'D' ,label = 'Square')
plt.plot(x,y_cu,'-o', label = 'Cube')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Polynomial Graph")
plt.legend()  #shows the legends
plt.tight_layout() #use if you have padding issues
plt.grid(True)
#plt.savefig('plot.png')
plt.show()
```

```python
width = 0.25
plt.bar(x-width,y_li, color = 'orange', width = width, label = 'Linear')
plt.bar(x,y_sq, color = 'red', width = width, label = 'Square')
plt.bar(x+width,y_cu, width = width, label = 'Cube')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Polynomial Graph")
# plt.xticks(ticks = x_indexes, labels = x)  it is used if you use x in respect of some other quantity {0,1,2} -> {25,26,27}, so ticks = {0,1,2} labels = {25,26,27}
plt.legend(loc = 0, prop = {'size':12})  #loc = location of graph| plt.legend(): shows the legend
plt.show()
```

```python
data = pd.read_csv("data.csv")
data.head()
```

```python
languages = data.iloc[:,-1].values
```

```python
from collections import Counter
count = Counter()
[count.update(y.split(';')) for y in languages]

first_15 = count.most_common(15)

langs = []
popularity = []
for x in first_15:
    langs.append(x[0])
    popularity.append(x[1])
langs = langs[::-1]
popularity = popularity[::-1]
```

```python
plt.figure(1, facecolor = 'white', figsize = (12.0,6.0))
plt.barh(langs, popularity)  #horizontal bargraph
plt.title("Most Popular Languages")
#plt.ylabel("Programming Languages")
plt.xlabel("No. of People who Use")
plt.tight_layout()
plt.savefig('Popular Programming Languages.png')
plt.show()
```
